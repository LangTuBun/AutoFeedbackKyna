```python
from selenium import webdriver
import tkinter as tk
from tkinter import messagebox
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import random
```

```python
def collect_user_input():
    # Variables to store user input
    user_info = {}

    # Function to handle the user input and close the window
    def submit_info():
        name = name_entry.get()
        password = password_entry.get()
        ngaydanhgia_1 = ngaydanhgia1.get()
        ngaydanhgia_2 = ngaydanhgia2.get()  
        
        
        
        if not name or not password:
            messagebox.showerror("Input Error", "Please fill in all fields")
        else:
            user_info['name'] = name
            user_info['password'] = password
            user_info['date_1'] = ngaydanhgia_1
            user_info['date_2'] = ngaydanhgia_2
            root.quit()  # Close the GUI loop

    # Create the main window
    root = tk.Tk()
    root.title("Nhap tai khoan mat khau vao day")
    root.geometry("400x200") 

    # Create and place the labels and entry widgets
    tk.Label(root, text="Name:").grid(row=0, column=0)
    name_entry = tk.Entry(root)
    name_entry.grid(row=0, column=3)

    tk.Label(root, text="Password:").grid(row=1, column=0)
    password_entry = tk.Entry(root)
    password_entry.grid(row=1, column=3)
    tk.Label(root, text="Ngay Danh Gia 1").grid(row=2, column=0)
    ngaydanhgia1 = tk.Entry(root)
    ngaydanhgia1.grid(row=2, column=3)
    
    tk.Label(root, text="Ngay Danh Gia 2").grid(row=3, column=0)
    ngaydanhgia2 = tk.Entry(root)
    ngaydanhgia2.grid(row=3, column=3)

    submit_button = tk.Button(root, text="Submit", command=submit_info)
    submit_button.grid(row=4, column=0, columnspan=2)
    root.mainloop()
    root.destroy()  # Properly close the window
    return user_info    
user_data = collect_user_input()
```

```python
url = 'https://teacher.kynaenglish.com/classes'
driver = webdriver.Edge()
driver.maximize_window()
driver.get(url)
```
